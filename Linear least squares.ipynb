```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([-1, 0, 1, 2])
y = np.array([0, -1, 2, 9])

n = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
c = (sum_y - m * sum_x) / n

print(f"g(x) = {m:.2f}x + {c:.2f}")
```

```python
plt.plot(x, y, 'o', label='datas', markersize=5)
plt.plot(x, m * x + c, 'r', label='line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 1.25, 1.50, 1.75, 2])
y = np.array([5.10, 5.79, 6.53, 7.45, 8.40])

log_y = np.log(y)

n = len(x)

sum_x = np.sum(x)
sum_log_y = np.sum(log_y)
sum_x_log_y = np.sum(x * log_y)
sum_x2 = np.sum(x ** 2)

b = (n * sum_x_log_y - sum_x * sum_log_y) / (n * sum_x2 - sum_x ** 2)
log_a = (sum_log_y - b * sum_x) / n
a = np.exp(log_a)

print(f"g(x) = {a:.4f}e^({b:.4f}x)")

x_fit = np.linspace(1, 2, 100)
y_fit = a * np.exp(b * x_fit)

plt.plot(x, y, 'o', label='datas', markersize=5)
plt.plot(x_fit, y_fit, 'r', label='line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
